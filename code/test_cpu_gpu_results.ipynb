```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/College/EdgeComputing/build/Debug/grid_large.csv', header=None)
df
```

```python
new_df = pd.read_csv('C:/College/EdgeComputing/build/Debug/expanded_interpolated_grid.csv', header=None)
new_df
```

```python
import matplotlib.pyplot as plt
```

```python
# Convert the DataFrame to a NumPy array.
elevation_matrix = df.values

# Create the heatmap.
plt.figure(figsize=(8, 6))
plt.imshow(elevation_matrix, cmap='viridis', origin='lower')
plt.colorbar(label='Elevation (m)')
plt.title("Bathymetry Heatmap")
plt.xlabel("Grid X")
plt.ylabel("Grid Y")
plt.show()
```

```python
# Convert the DataFrame to a NumPy array.
new_elevation_matrix = new_df.values

# Create the heatmap.
plt.clf()
plt.figure(figsize=(8, 6))
plt.imshow(new_elevation_matrix, cmap='viridis', origin='lower')
plt.colorbar(label='Interpolated Elevations (m)')
plt.title("Bathymetry Heatmap")
plt.xlabel("Grid X")
plt.ylabel("Grid Y")
plt.show()
```
