```python
orig_elevation_matrix = og.values
```

```python
orig_elevation_matrix.shape
```

### CPU Bilinear

```python
cpu_bilin = pd.read_csv(r'.\interpolated_cpu.csv')
```

```python
cpu_bilin['interpolated_value'].values
```

```python
# Reconstruct the grid with the interpolated values included
missing = pd.read_csv(r'.\reference_missing.csv', header=None)

updates = np.array(missing.values)
updates
```

```python
orig_elevation_matrix.shape
```

```python
updates.shape
```

```python
for x, y, val in updates:
    x = int(x); y = int(y)   # ensure integer indices
    # optional bounds check
    if 0 <= x <= orig_elevation_matrix.shape[0] and 0 <= y <= orig_elevation_matrix.shape[1]:
        orig_elevation_matrix[x, y] = val
    else:
        raise IndexError(f"Index out of range: ({x},{y})")
```

```python
# Convert the DataFrame to a NumPy array.
new_elevation_matrix = cpu_bilin['interpolated_value'].values.reshape

# Create the heatmap.
plt.clf()
plt.figure(figsize=(8, 6))
plt.imshow(orig_elevation_matrix, cmap='viridis', origin='lower')
plt.colorbar(label='Original Elevations (m)')
plt.title("Original Grid")
plt.xlabel("Grid X")
plt.ylabel("Grid Y")
plt.show()

# Create the heatmap.
plt.clf()
plt.figure(figsize=(8, 6))
plt.imshow(new_elevation_matrix, cmap='viridis', origin='lower')
plt.colorbar(label='Interpolated Elevations (m)')
plt.title("CPU Bilinear Interpolation (GEBCO)")
plt.xlabel("Grid X")
plt.ylabel("Grid Y")
plt.show()
```
