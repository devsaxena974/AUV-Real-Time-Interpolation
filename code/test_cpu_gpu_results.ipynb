```python
def fill_reduced_grid(reduced_grid, interp_values, update_indices):
    interp_count = 0
    for entry in update_indices:
        r = int(entry[0])
        c = int(entry[1])

        
        reduced_grid.iloc[r, c] = interp_values[interp_count]
        interp_count += 1
        
    has_nan = reduced_grid.isnull().values.any()

    if has_nan:
        print("The DataFrame contains NaN values.")
        return ValueError()
    else:
        print("Succesfully filled in reduced grid")
        return 0
```
